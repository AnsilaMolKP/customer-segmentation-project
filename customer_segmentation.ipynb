```python
import pandas as pd
```

```python
df = pd.read_csv("C:/Users/ANSILA/Downloads/store_customers.csv")
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df = df.dropna()
```

```python
df['Gender'] = df['Gender'].str.strip().str.capitalize()
```

```python
#drop id
df = df.drop('CustomerID', axis=1)
```

```python
#feature selection
X = df[['Age', 'Annual Income (k$)', 'Spending Score (1-100)']]
```

```python
#apply k means
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=5, random_state=42)
df['Cluster'] = kmeans.fit_predict(X)
```

```python
df.columns = ['Gender', 'Age', 'Income', 'SpendingScore', 'Cluster']
```

```python
df.to_excel("clean_data.xlsx", index=False)
```
